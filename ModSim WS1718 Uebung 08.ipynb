````python
%matplotlib notebook
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
````

````python
Ltau = 200                                               #legt die Anzahl an Punkten auf 200 fest
Lpsi = 200                                               #legt die Anzahl an Punkten auf 200 fest
r = 5                                                    #legt den Radius auf 5 fest
tau = np.linspace(0,pi,Ltau + 1)                         #erzeugt eine 1xLtau+1 Matrix 'tau' im Intervall von 0 bis pi mit Ltau+1 Werten
psi = np.linspace(0,2 * pi,Lpsi +1)                      #erzeugt eine 1xLpsi+1 Matrix 'tau' im Intervall von 0 bis pi mit Lpsi+1 Werten

TAU,PSI = np.meshgrid(tau,psi)                           #erzeugt aus den Arrays ein zweidimensionales Koorinatensystem mit den Werten von 'tau' auf der x-achse und den Werten von 'psi' auf der y-Achse

x = r * sin(TAU) * cos(PSI)                              #Gleichung zur Parametrisierung der x-Werte
y = r * sin(TAU) * sin(PSI)                              #Gleichung zur Parametrisierung der y-Werte
z = r * cos(TAU)                                         #Gleichung zur Parametrisierung der z-Werte

fig = figure('3D function wireframe plot')               #Figur soll ein 3D Figur sein
ax = fig.add_subplot(111,projection='3d')   

######### ax3d=fig.add_subplot(111,projection='3d',aspect=1) mit aspect =1 wird der Kugel sphÃ¤risch aussiehen

wf = ax.plot_wireframe(x,y,z,rstride=10,cstride=10)      #definiert die zu verwendenen Parameter fÃ¼r den Plot

ax.set_xlabel('$x$',fontsize=10)                         #setze Name der x-Achse auf 'x'
ax.set_ylabel('$y$',fontsize=10)                         #setze Name der y-Achse auf 'y'
ax.set_zlabel('$z$',fontsize=10)                         #setze Name der z-Achse auf 'z'
ax.set_title('$Kugel$')                                  #setze 'Kugel' als Titel der Grafik
show()
````

````python
#+10
def Tknot(p,q,R=2,r=1,h=0.1,Lp=200,Lq=200):                         #definiert die Funktion Tknot mit den gegebenen Parametern
    phi = np.linspace(0,2*pi,200)                                   #erzeugt ein Array 'phi' von '0' bis '2pi'mit 200 Werten
    th = np.linspace(0,2*pi,200)                                    #erzeugt ein Array 'th' von '0' bis '2pi'mit 200 Werten

    PHI,TH = np.meshgrid(phi,th)                                    #erzeugt aus den Arrays ein zweidimensionales Koorinatensystem mit den Werten von 'phi' auf der x-achse und den Werten von 'th' auf der y-Achse
    
    x = (R+r*cos(p*PHI))*cos(q*PHI)                                 #Gleichung zur Parametrisierung der x-Werte                              
    y = (R+r*cos(p*PHI))*sin(q*PHI)                                 #Gleichung zur Parametrisierung der y-Werte   
    z = r*sin(p*PHI)                                                #Gleichung zur Parametrisierung der z-Werte   
    
    fig = figure()
    ax3d = fig.add_subplot(111,projection='3d')
    ax3d.plot_wireframe(x,y,z,rstride=10,cstride=10, linewidth = h) #erstellt den Torusknoten mit den ober bestimmten Parametern

    torus = ax3d.plot_surface((R+r*cos(TH))*cos(PHI),(R+r*cos(TH))*sin(PHI),r*sin(TH),rstride=10,cstride=10,linewidth=1,alpha=0.2)
    #erzeugt das Torusband um den Torusknoten mit den oben bestimmten Parametern
    
    ax.set_xlim(-(R+r+0.5),R+r+0.5)                                #begrenzt die x-Achse von '-(R+r)' bis (R+r)'
    ax.set_ylim(-(R+r+0.5),R+r+0.5)                                #begrenzt die y-Achse von '-(R+r)' bis (R+r)'
    ax.set_zlim(-(R+r+0.5),R+r+0.5)                                #begrenzt die z-Achse von '-(R+r)' bis (R+r)'
    ax.set_xlabel('$x$')                                           #setze Name der x-Achse auf 'x'
    ax.set_ylabel('$y$')                                           #setze Name der y-Achse auf 'y'
    ax.set_zlabel('$z$')                                           #setze Name der z-Achse auf 'z'
    ax.set_title('$Torusknoten$')                                  #setze 'Torusknoten' als Titel der Grafik
    show()
````

- Schreiben Sie eine Funktion `Newton_Steps(x0,f,fprime,dmin=1e-15,dmax=1e3,Nmax=100)` welche das Newton Verfahren mit der Funktion `f(x)` und deren Ableitung `fprime(x)` beginnend bei $x_0$ solange ausfuehrt bis $\left|f(x_n)\right|\le d_{min}$, $\left|f(x_n)\right|\ge d_{max}$ oder $n=N_{max}$ und den Wert $x_n$ der Nullstellen-Naeherung sowie die Zahl $n$ der notwendigen Iterationsschritte ausgibt.

- Bestimmen Sie fuer alle *komplexen* Anfangswerte $z =x+iy $ auf einem geeigneten Gitter die Approximation der von dort gefundenen Nullstellen der Funktion $f(z)=z^3-1$ und die Zahl der benoetigten Iterationsschritte. Hinweis : Wenn Sie die Funktionen `Newton_Steps`, `f` und `fprime` sorgfaeltig programmieren, koennen diese direkt auf ein ganzes `numpy` Feld von Anfangswerten gleichzeitig angewendet werden ohne Python Schleifen zu verwenden.

#### Zusatzaufgabe :

- Definieren Sie eine Funktion `polymaker(c)` welche bei einem gegebenen iterierbaren Container `c` mit Nullstellen $c_i$ die Funktion `f(x)`$=\prod_i (x-c_i)$ und deren Ableitung `fprime(x)`$=\sum_i \prod_{j\ne i} (x-c_j)$ dynamisch generiert und als Funktionsobjekte zurueckliefert.

- Plotten Sie Beispiele fuer Newton Fraktale von verschiedenen Funktionen, die Sie mit `polymaker(c)` erzeugen.

````python
def Newton_Steps(x0,f,fprime,dmin=1e-15,dmax=1e3,Nmax=100):          #definiert eine Funktion Newton_Steps mit den gegebenen Parametern
    n = 0                                                            #setzte n = 0
    nullx = x0                                                       #definiert die Nullstelle als x0
    nully = f(nullx)                                                 #definiert nully als Funktion an der Stelle x0 
    while abs(nully) >= dmin or abs(nully) <= dmax or n <= 50:       #solange eine der folgenden Bedingungen erfÃ¼llt ist fÃ¼hre die Schleife aus
                                                                     #1. Betrag von nully grÃ¶ÃŸer gleich als dmin ist
                                                                     #2. Betrag von nully ist kleiner gleich dmin
                                                                     #oder 3. n kleiner gleich 50 ist ################ statt n <= 50: n <= Nmax -2
        nullx = nullx - (f(nullx)/fprime(nullx))                     #berechne nullx 
        nully = f(nullx)                                             #berechne nully erneut mit dem neuen nullx
        n = n + 1                                                    #erweitere n mit 1
        
    return (nullx,n)                                                 #gebe n und nullx als Ergebnisse der Funktion aus

#-25 zweite Teil
````
